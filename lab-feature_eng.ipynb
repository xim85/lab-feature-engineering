There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
final_spaceship= spaceship.dropna()
final_spaceship
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
final_spaceship['Cabin'] = final_spaceship['Cabin'].apply(lambda x: x.split('/')[0][0] if pd.notnull(x) else 'Unknown')
final_spaceship
```

```python
final_spaceship['Cabin'].unique()
```

- Drop PassengerId and Name

```python
#your code here
final_spaceship.drop(columns=['PassengerId', 'Name'], inplace=True)
final_spaceship
```

- For non-numerical columns, do dummies.

```python
#your code here
X = final_spaceship.drop(columns="Age")
y = final_spaceship["Age"]
```

```python
X_num = X.select_dtypes(include='number')
X_cat = X.select_dtypes(include='object')

X_num.shape , X_cat.shape
```

```python
encoder = OneHotEncoder().fit(X_cat) 
encoded = encoder.transform(X_cat).toarray()
unique_values = [encoder.categories_[i] for i in range(len(X_cat.columns))]
onehot_encoded = pd.DataFrame(encoded, columns=[f"{col}_{value}" for col, values in zip(X_cat.columns, unique_values) for value in values]) 
onehot_encoded
```

```python
onehot_encoded.reset_index (drop=True, inplace = True)
X_num.reset_index (drop=True, inplace=True)
```

```python
X = pd.concat([X_num, onehot_encoded], axis=1)
X
```

**Perform Train Test Split**

```python
#your code here
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

```python
transformer = MinMaxScaler().fit(X_train) #Aplicar el min y el max, pero recuerda que X  tiene que ser valores numericos
X_train_normalized_array = transformer.transform(X_train) #Convertirlo entre 0 y 1 
X_train_norm= pd.DataFrame(X_train_normalized_array, columns=X_train.columns) 
```

```python
X_test_normalized_array = transformer.transform(X_test) #Convertirlo entre 0 y 1 
X_test_norm = pd.DataFrame(X_test_normalized_array, columns=X_test.columns) 

X_test_norm
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here
knn_norm = KNeighborsRegressor(n_neighbors=10)
```

```python
knn_norm.fit(X_train_norm, y_train)
knn_norm_score = knn_norm.score(X_test_norm, y_test)
```

```python
knn_norm_score
```

```python
from sklearn.metrics import r2_score
```

```python
predictions_norm = knn_norm.predict(X_test_norm)
r2_norm = r2_score(y_test, predictions_norm)

r2_norm 
```
